```python
import matplotlib.pyplot as plt
import numpy as np #for randomly generating numbers

x = np.arange(10)
y1 = x**2
y2 = x**3
data1 = np.random.randint(1, 10, 50)
data2 = np.random.randint(2, 11, 51)
data3 = np.random.randint(3, 12, 55)
data4 = np.random.randint(4, 15, 60)

# Step 1: Subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))#2 rows and 2 columns , 12 is width and 10 is height

# Line plot
axs[0, 0].plot(x, y1, label='x^2')#axs[0,0], first row and first column
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].legend()


# Scatter plot
axs[0, 1].scatter(x, y2, color='r', label='x^3')#first row and second column
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].legend()

# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
axs[1, 0].bar(categories, values, color='g')#second row and first column
axs[1, 0].set_title('Bar Plot')
axs[1, 0].set_xlabel('Categories')
axs[1, 0].set_ylabel('Values')


# Histogram
axs[1, 1].hist(data1, bins=5, alpha=0.7, label='Data1')#second row and second column
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].legend()

plt.tight_layout()#to adjust subplot parameters so that sub plots fit into figure area
plt.show()
print('\n')
# Step 2: Histograms
plt.figure(figsize=(10, 6))

# Histogram for a single dataset
plt.hist(data1, bins=5, alpha=0.5, label='Data1', color='b')#bins specifies the number of intervals of histogram and alpha sets the transparency level of histogram i.e 0 value means more transparent 
plt.hist(data2, bins=5, alpha=0.5, label='Data2', color='r')
plt.title('Multiple Histograms')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
print('\n')

# Normalized histograms
plt.figure(figsize=(10, 6))
plt.hist(data1, bins=5, alpha=0.5, label='Data1', density=True, color='b')#density=true represents normalization of histogram
plt.hist(data2, bins=5, alpha=0.5, label='Data2', density=True, color='r')
plt.title('Normalized Histograms')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
print('\n')
# Step 3: Box Plots
plt.figure(figsize=(10, 6))

# Single box plot
plt.boxplot(data1)
plt.title('Box Plot for Data1')
plt.ylabel('Value')
plt.show()
print('\n')
# Multiple box plots
plt.figure(figsize=(10, 6))
plt.boxplot([data1, data2, data3, data4], patch_artist=True,#allows filling the box plot boxes with colors
            boxprops=dict(facecolor='cyan', color='blue'),#facecolor means color inside the box will be cyan, and color=blue means the outline color will be blue
            medianprops=dict(color='red'))#median line color inside the box is red
plt.title('Multiple Box Plots')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.xticks([1, 2, 3, 4], ['Data1', 'Data2', 'Data3', 'Data4'])
plt.show()



```
