```python
import pandas as pd

# Load a DataFrame from a CSV file
df = pd.read_csv(r'C:\Users\Huawei\Desktop\PrediveAIModelForPakistanStockExchange\350PakistanStocksExchangeStocksData\abl(AlliedBankLtd).csv')  

# Display the first and last five rows of the DataFrame
print("Display the first and last five rows of the DataFrame\n")
print(df.head())  # display the first five rows
print(df.tail())  # display the last five rows
print("\n");

# Set a specific column as the index of the DataFrame
df.set_index('TIME', inplace=True)  

# Select a specific column and display its values
print("Select a specific column and display its values\n")
print(df['CLOSE'])  
print("\n");
# Select multiple columns and display the resulting DataFrame
print("Select multiple columns and display the resulting DataFrame\n")
print(df[['HIGH', 'LOW']])  
print("\n");
# Select a subset of rows using the .loc method
print("Select a subset of rows using the .loc method\n")
print(df.loc['9-Oct-20'])  #select a row
print(df.loc['9-Oct-20':'6-Oct-20'])  # select a range of rows
print("\n");
# Select a subset of rows and columns using the .iloc method
print("Select a subset of rows and columns using the .iloc method\n")
print(df.iloc[0])  # select the first row
print(df.iloc[0:3, 1:4])  # select a subset of rows and columns
print("\n");
# Filter rows based on a condition
filtered_df = df[df['VOLUME'] > 50000] 
print("Filter rows based on a condition\n")
print(filtered_df)
print("\n");
# Group the DataFrame by a specific column and calculate the mean of each group
grouped_mean = df.groupby('OPEN').mean() 
print("Group the DataFrame by a specific column and calculate the mean of each group\n")
print(grouped_mean)
print("\n");
# Group the DataFrame by multiple columns and calculate the sum of each group
grouped_sum = df.groupby(['OPEN', 'HIGH']).sum()  
print("Group the DataFrame by multiple columns and calculate the sum of each group\n")
print(grouped_sum)
print("\n");
# Use the agg method to apply multiple aggregation functions to grouped data
agg_functions = df.groupby('OPEN').agg({'HIGH': 'mean', 'LOW': 'sum'})  
print("Use the agg method to apply multiple aggregation functions to grouped data\n")
print(agg_functions)
print("\n");
# Calculate the size of each group
group_size = df.groupby('CLOSE').size() 
print("Calculate the size of each group\n")
print(group_size)
print("\n");
# Select rows based on multiple conditions
filtered_rows = df[(df['HIGH'] > 85) & (df['LOW'] < 83)] 
print("Select rows based on multiple conditions\n")
print(filtered_rows)
print("\n");
# Use the query method to filter rows
queried_df = df.query('OPEN > 84') 
print("Use the query method to filter rows\n")
print(queried_df)
print("\n");
# Use isin to filter rows based on a list of values
filtered_by_list = df[df['CLOSE'].isin(['83', '85', '86'])]  
print("Use isin to filter rows based on a list of values\n")
print(filtered_by_list)
print("\n");
# Select specific columns and rename them
renamed_df = df[['HIGH', 'LOW']].rename(columns={'HIGH': 'highPriceOfStock', 'LOW': 'lowPriceOfStock'})  
print("Select specific columns and rename them\n")
print(renamed_df)
```
