```python
import matplotlib.pyplot as plt

#Basic Plots
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
y2 = [1,4,6,8,10]
#1. Simple Line Plot
plt.plot(x, y)
plt.show()


#2. Multiple Lines on the Same Graph
plt.plot(x, y)
plt.plot(x, y2)
plt.show()

#3. Adding Markers to a Line Plot
plt.plot(x, y, marker='s')#s for square shape marker, it can also be o where o means circle shaped marker
plt.show()

#4. Customizing Line Styles and Colors
plt.plot(x, y, linestyle='-.', color='b')#line styles can be --, - , b in color means blue
plt.show()

#Scatter Plots
#1. Basic Scatter Plot
plt.scatter(x, y)
plt.show()

#2. Customizing Marker Styles, Sizes, and Colors
plt.scatter(x, y, marker='D', s=100, c='green')
plt.show()

#3. Plotting Multiple Scatter Plots on the Same Graph
plt.scatter(x, y)
plt.scatter(x, y2)
plt.show()

#4. Adding Labels and Titles to the Scatter Plot
plt.scatter(x, y)
plt.title('Scatter Plot with Labels')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

#Bar Charts
#1. Vertical Bar Chart
P=['A','B','C','D']
M=[1,2,3,4]
plt.bar(P, M)
plt.show()

#2. Horizontal Bar Chart
plt.barh(P, M)
plt.show()

#3. Customizing Bar Colors and Adding Edge Colors
plt.bar(P, M, color='purple', edgecolor='black')
plt.show()

#Pie Charts
#1. Basic Pie Chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels)
plt.show()

#2. Customizing Slice Colors and Adding a Legend
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors)
plt.legend(labels)
plt.show()

#3. Exploding a Slice of the Pie Chart
explodes = (0, 0.2, 0, 1)#explode the 2nd and 4th slice only
plt.pie(sizes, labels=labels, colors=colors, explode=explodes)
plt.show()

#4. Adding Percentage Labels to Slices
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.03f%%')
#%for formatting the percentage
#f is for floating point numbers
#1 is total width of percentage string, including the decimal point
#.03 indicates 3 decimal places after point
#%% to print percentage sign
plt.show()
```
