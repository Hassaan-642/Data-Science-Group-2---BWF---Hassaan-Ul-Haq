```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Huawei\Desktop\PrediveAIModelForPakistanStockExchange\350PakistanStocksExchangeStocksData\abl(AlliedBankLtd).csv')
df1 = pd.read_csv(r'C:\Users\Huawei\Desktop\PrediveAIModelForPakistanStockExchange\350PakistanStocksExchangeStocksData\akbl(AskariBankLtd).csv')
df2 = pd.read_csv(r'C:\Users\Huawei\Desktop\PrediveAIModelForPakistanStockExchange\350PakistanStocksExchangeStocksData\abot(AbbotLabortariesLtd).csv')
df3 = pd.read_csv(r'C:\Users\Huawei\Desktop\PrediveAIModelForPakistanStockExchange\350PakistanStocksExchangeStocksData\acpl(AttockCementPakistanLtd).csv')
df4 = pd.read_csv(r'C:\Users\Huawei\Desktop\PrediveAIModelForPakistanStockExchange\350PakistanStocksExchangeStocksData\adams(AdamSugarMillsLimited).csv')
df5 = pd.read_csv(r'C:\Users\Huawei\Desktop\PrediveAIModelForPakistanStockExchange\350PakistanStocksExchangeStocksData\admm(ArtisticDenimMillsLimited).csv')

#Merge two DataFrames on a single key.
print('Merge two DataFrames on a single key.\n')
merged_df=pd.merge(df,df1, on='CLOSE', suffixes=('abl','akbl'))#suffixes for separating two files attributes and on is used on which basis dataframes are merged
print(merged_df)
#Merge two DataFrames on multiple keys.
print('\nMerge two DataFrames on multiple keys.\n')
merged_df2=pd.merge(df2,df3, on=['HIGH','LOW'], suffixes=('abot','acpl'))
print(merged_df2)
#Perform an outer join, inner join, left join, and right join.
print('\nPerform an outer join, inner join, left join, and right join..\n')
print('innerJoin\n')
innerJoin=pd.merge(df4,df5, how='inner', on='CLOSE', suffixes=('adams','admm'))#suffixes for separating two files attributes
print(innerJoin)
print('\nouterJoin\n')
outerJoin=pd.merge(df4,df5, how='outer', on='HIGH', suffixes=('adams','admm'))#suffixes for separating two files attributes
print(outerJoin)
print('\nLeftJoin\n')
leftJoin=pd.merge(df4,df5, how='left', on='LOW', suffixes=('adams','admm'))#suffixes for separating two files attributes
print(leftJoin)
print('\nrightJoin\n')
rightJoin=pd.merge(df4,df5, how='right', on='OPEN', suffixes=('adams','admm'))#suffixes for separating two files attributes
print(rightJoin)
#Concatenate two DataFrames along rows.
concat_df=pd.concat([df,df1],axis=0)#axis=0 means concatenate along rows
print('\nConcatenate two DataFrames along rows.\n')
print(concat_df)
#Concatenate two DataFrames along columns.
concat_df2=pd.concat([df2,df3],axis=1)#axis=1 means concatenate along columns
print('\nConcatenate two DataFrames along columns.\n')
print(concat_df2)
#Concatenate a list of DataFrames
df_list=[df,df3]
concate_df=pd.concat(df_list)
print('\nConcatenate a list of DataFrames\n')
print(concate_df)
#Reshape data using the melt function
print('\nReshape data using the melt function\n')
melted_df = pd.melt(df, id_vars=['HIGH'], value_vars=['LOW', 'CLOSE'])#is_vars=HIGH represents that HIGH will remain fixed and value_vars=LOW,CLOSE represents that it will be melted and will be transformed into a longer format. 
print(melted_df)
#Create a pivot table to summarize data
print('\nCreate a pivot table to summarize data\n')
pivot_table = pd.pivot_table(df, values='CLOSE', index=['HIGH', 'LOW'], columns=['VOLUME'], aggfunc='sum')
#values specifiy the column whose values will be aggregated, i.e close
#index=HIGH and LOW will be used as the index of the pivot table. This means the rows of the Pivot table will be uniquely defined by combination of HIGH and LOW values.
#columns_this column unique values will become columnn of the pivot table
#agg_function to be used is sum
print(pivot_table)
print('\n')
#Group data by one or more columns and perform aggregation functions
print('Group data by one or more columns and perform aggregation functions\n')
grouped = df.groupby('HIGH').agg({'LOW': ['sum', 'mean', 'count']})#group the data by high column, similar high values will be grouped and according to that high column group, Low values (sum, mean and count) will be caluclated for each group
print(grouped)
#Apply multiple aggregation functions to grouped data
print('\nApply multiple aggregation functions to grouped data\n')
grouped = df.groupby('CLOSE').agg({'HIGH': ['sum', 'mean', 'count'], 'LOW': 'count'})#data frame is grouped beased on unique values of Close Columns.
print(grouped)
#Use the groupby function to group data and apply custom functions
print('\nUse the groupby function to group data and apply custom functions\n')
grouped = df.groupby('CLOSE').apply(lambda x: x[x['LOW'] > 50])
#dataframe is grouped based on unique values  in the close column
#the lambda function filters rows with in each group(' x') where the value in the low column is greater than 50
print(grouped)


```
