#Scatter Plot

```python
import pandas as pd
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
df = pd.read_csv(r'C:\Users\Huawei\Desktop\PrediveAIModelForPakistanStockExchange\KSE100_PakistanStocksExchangeStocksData\abl(AlliedBankLtd).csv')


sns.set(style="whitegrid")
scatter_plot = sns.scatterplot(data=df, x="OPEN", y="CLOSE")
scatter_plot.set(title="Scatter Plot: Open vs Close Prices", xlabel="Open Price", ylabel="Close Price")
```

#Line Plot

```python
line_plot = sns.lineplot(data=df, x="HIGH", y="LOW")
line_plot.set(title="Line Plot: HIGH Price Over LOW", xlabel="HIGH", ylabel="LOW")
```

#Bar Plot

```python

bar_plot = sns.barplot(data=df, x="CLOSE", y="LOW")
bar_plot.set(title="Bar Plot: Trading Close Over LOW", xlabel="Close", ylabel="Low")
```

#Count Plot

```python
sns.set_style("whitegrid")
countplot = sns.countplot(x='CLOSE', data=df)
countplot.set(title="Count Plot: Close Count")
```

#Box Plot

```python

box_plot = sns.boxplot(data=df[['OPEN', 'HIGH', 'LOW', 'CLOSE']])
box_plot.set(title="Box Plot: Price Distribution", xlabel="Price Type", ylabel="Price")
```

#Customization

#Setting the Style use: sns.set_style()

```python
sns.set_style("ticks")  
bar_plot = sns.barplot(data=df, x="OPEN", y="VOLUME")
bar_plot.set(title="Setting the Style use: sns.set_style()")
```

#Adding Titles and Labels

```python
line_plot = sns.lineplot(data=df, x="OPEN", y="HIGH")
line_plot.set(title="Adding Titles and Labels: Open Price Over High", xlabel="Open", ylabel="High")
```

#Changing Palette

```python
sns.set_palette("bright")  
bar_plot = sns.barplot(data=df, x="OPEN", y="HIGH")
bar_plot.set(title="Changing Palette: Open Price vs High ")
```
